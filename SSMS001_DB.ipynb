```python
# calculate number of employees and minimum and maximum salary and average of each department

try:
    conn = pyodbc.connect(connection_string)
    print("✅ Connected successfully!")

    cursor = conn.cursor()

    # Execute SELECT query
    query = "select d.DNAME as dname, count(e.empno) as nbr_emp, Max(e.sal) as Sal_max, Min(sal) as sal_min, avg(sal) as sal_moy from DEPT d inner join emp e on d.DEPTNO = e.DEPTNO group by d.DNAME order by Max(e.sal) desc ;"
    cursor.execute(query)
    print(f"✅ Executed: {query.strip()}")

    # Fetch and display results
    records = cursor.fetchall()
    for r in records:
        print(f"{r.dname}\t{r.nbr_emp}\t{r.Sal_max}\t{r.sal_min}\t{r.sal_moy}")

    print("✅ All rows displayed successfully.")

    # Convert to NumPy array
    data_np_03 = np.array([list(row) for row in records])
    print("📦 NumPy Array:")
    print(data_np_03)

    # Convert to Pandas DataFrame
    df_03 = pd.DataFrame(data_np_03, columns=["Name Departement", "Nbr emp", "Max Sal","Min Sal", "Moy Sal"])
    print("\n📊 Pandas DataFrame:")
    print(df_03)

    conn.close()

except Exception as e:
    print("❌ Connection or execution error:", e)
```

```python
import pandas as pd

# Step 1: Rename columns
df_emp = df_emp.rename(columns={"N°Dept": "DEPTNO"})
df_dept = df_dept.rename(columns={"N°departement": "DEPTNO", "Name departement": "DNAME"})

# Step 2: Clean and convert DEPTNO to ensure perfect join keys
df_emp["DEPTNO"] = df_emp["DEPTNO"].astype(str).str.strip()
df_dept["DEPTNO"] = df_dept["DEPTNO"].astype(str).str.strip()

# Step 3: Convert salary to numeric
df_emp["Sal"] = pd.to_numeric(df_emp["Sal"], errors="coerce")

# Step 4: Merge DataFrames
merged_df = pd.merge(df_emp, df_dept, on="DEPTNO", how="inner")

# Step 5: Group and aggregate
result_df = merged_df.groupby("DNAME").agg(
    Nbr_emp=("N°employe", "count"),
    Sal_max=("Sal", "max"),
    Sal_min=("Sal", "min"),
    Sal_moy=("Sal", "mean")
).reset_index()

# Step 6: Sort by max salary
result_df = result_df.sort_values(by="Sal_max", ascending=False)

# Show result
print(result_df)
```
