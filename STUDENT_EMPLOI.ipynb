```python
import pandas as pd
```

```python
# Define days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
print(days)
```

```python
# Define time slots
time_slots = [(8, 9), (9, 10), (10, 11), (11, 12), (12, 13), (13, 14), (14, 15), (15, 16)]
```

```python
print(time_slots)
```

```python
# Define students
students = ['Mario', 'Luigi', 'Kango', 'Bowser']
```

```python
print(students)
```

```python
# List of available courses
course_list = ["SQL", "Python", "R", "Linear Algebra", "Statistics", "Probability",
               "Machine learning", "Deep learning", "Power BI"]
```

```python
print(course_list)
```

```python
# Create MultiIndex columns: First level student names, second level time slots
columns = pd.MultiIndex.from_product([students, [f"{t[0]},{t[1]}" for t in time_slots]], names=['Student', 'Time'])
```

```python
print(columns)
```

```python
# Initialize empty DataFrame with days as index and MultiIndex columns
schedule_df = pd.DataFrame(index=days, columns=columns)
```

```python
print(schedule_df)
```

```python
# Function to add a course to a student's schedule
def add_course(student, day, time_slot, course):
    # Find the correct column
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    schedule_df.at[day, col] = course
```

```python
# Function to remove a course from a student's schedule
def remove_course(student, day, time_slot):
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    schedule_df.at[day, col] = None
```

```python
# Function to modify a course in a student's schedule
def modify_course(student, day, time_slot, new_course):
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    schedule_df.at[day, col] = new_course
```

```python
# Example: Adding courses for students
add_course('Mario', 'Monday', (8, 9), 'SQL')
add_course('Luigi', 'Monday', (9, 10), 'Python')
add_course('Kango', 'Wednesday', (10, 11), 'R')
add_course('Bowser', 'Thursday', (11, 12), 'Algebre lineaire')
```

```python
# Display schedule
print("Initial Schedule:\n")
display(schedule_df)
```

```python
# Updated add_course function with subject validation and error messages
def add_course(student, day, time_slot, course):
    if course not in course_list:
        print("❌ Error: This subject is not available in the school.")
        return
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    # Check if the student already has a course at this time
    if pd.notna(schedule_df.at[day, col]):
        print("❌ Error: This course is already assigned to this student at this day and time.")
        return
    # Check if the slot is occupied by another student's course
    slot_columns = schedule_df[(student, f"{time_slot[0]},{time_slot[1]}")]
    if slot_columns.notna().any():
        print("❌ Error: This time slot is already occupied by another course.")
        return
    # Assign the course
    schedule_df.at[day, col] = course
    print(f"✅ Success: Course assigned to {student} on {day} at {time_slot[0]},{time_slot[1]}: {course}")
```

```python
# Updated remove_course function with validation
def remove_course(student, day, time_slot):
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    existing_course = schedule_df.at[day, col]
    if pd.isna(existing_course):
        print("❌ Error: No course assigned in this slot for this student.")
        return
    # Remove the course
    schedule_df.at[day, col] = None
    print(f"✅ Success: Course removed from {student} on {day} at {time_slot[0]},{time_slot[1]}.")
```

```python
# Updated modify_course function with validation
def modify_course(student, day, time_slot, new_course):
    if new_course not in course_list:
        print("❌ Error: This subject is not available in the school.")
        return
    col = (student, f"{time_slot[0]},{time_slot[1]}")
    existing_course = schedule_df.at[day, col]
    if pd.isna(existing_course):
        print("❌ Error: No course to modify in this slot for this student.")
        return
    # Check if the new course conflicts with another course in this slot
    slot_columns = schedule_df[(student, f"{time_slot[0]},{time_slot[1]}")]
    if pd.notna(slot_columns).any():
        print("❌ Error: This time slot is already occupied by another course.")
        return
    # Modify the course
    schedule_df.at[day, col] = new_course
    print(f"✅ Success: Course modified for {student} on {day} at {time_slot[0]},{time_slot[1]}: {new_course}")
```

```python
# Example: Adding courses for students
add_course('Mario', 'Monday', (10, 11), 'Python')
add_course('Luigi', 'Monday', (8, 9), 'SQL')
add_course('Kango', 'Wednesday', (12, 13), 'Python')
add_course('Bowser', 'Thursday', (10, 11), 'Python')
```

```python
while True:
    print("Welcome to my student course schedule program.")

    user_choice = ""

    # ✅ Include 'Q' to allow quitting the loop
    while user_choice.upper() not in ("A", "R", "M", "D", "Q"):
        user_choice = input(
            "\nPlease choose an option:\n"
            "👉 To display the students' course table, enter: 'D'\n"
            "➕ To add a course to a student, enter: 'A'\n"
            "➖ To remove a course from a student, enter: 'R'\n"
            "✏️ To modify a course for a student, enter: 'M'\n"
            "❌ To quit the program, enter: 'Q'\n"
            "Your choice: "
        )

        if user_choice.upper() not in ("A", "R", "M", "D", "Q"):
            print(f"❌ Invalid operation choice: {user_choice}. Please enter A, R, M, D or Q.")
        else:
            print(f"✅ Valid operation choice: {user_choice.upper()}")

    # Exit the loop if user chooses Q
    if user_choice.upper() == "Q":
        print("👋 Exiting the program. Goodbye!")
        break

    elif user_choice.upper() == "D":
        print(schedule_df)
        
        print("🔍 Displaying the course table...")
        name_etudiant = " "
        while name_etudiant.upper() not in [s.upper() for s in students]:
            name_etudiant = input(f"Please choose a student name from this list:\n{students}\n")
            for x in students:
                if name_etudiant.upper() == x.upper():
                    name_etudiant = x
        print(f"✅ Selected student name: {name_etudiant}")
        
        # Select only Luigi's schedule
        student_schedule = schedule_df[name_etudiant]

        # Display Luigi's schedule
        print(f"{name_etudiant}'s schedule:\n")
        display(student_schedule)
        

    elif user_choice.upper() in ("A", "R", "M"):
        # --------------------------------------
        name_etudiant = " "
        while name_etudiant.upper() not in [s.upper() for s in students]:
            name_etudiant = input(f"Please choose a student name from this list:\n{students}\n")
            for x in students:
                if name_etudiant.upper() == x.upper():
                    name_etudiant = x
        print(f"✅ Selected student name: {name_etudiant}")

        day_ = " "
        while day_.upper() not in [s.upper() for s in days]:
            day_ = input(f"Please choose a study day from this list:\n{days}\n")
            for x in days:
                if day_.upper() == x.upper():
                    day_ = x
        print(f"📅 Selected study day: {day_}")

        chosen_time = " "
        while chosen_time not in time_slots:
            chosen_time = input(f"Please choose a class time from this list:\n{time_slots}\n")
            try:
                res = tuple(map(int, chosen_time.strip("()").split(',')))
                chosen_time = res
            except:
                print("⚠️ Invalid time format. Please use format like (8, 10)")
            continue
        print(f"⏰ Selected time slot: {chosen_time}")
        # --------------------------------------

        if user_choice.upper() == "R":
            print("➖ Removing a course...")
            remove_course(name_etudiant, day_, chosen_time)

        elif user_choice.upper() == "A":
            cours_ = " "
            while cours_.upper() not in [s.upper() for s in course_list]:
                cours_ = input(f"Please choose a subject from this list:\n{course_list}\n")
                for x in course_list:
                    if cours_.upper() == x.upper():
                        cours_ = x
            print(f"📘 Selected subject: {cours_}")
            print("➕ Adding the course...")
            add_course(name_etudiant, day_, chosen_time, cours_)

            # Select only Luigi's schedule
            student_schedule = schedule_df[name_etudiant]

            # Display Luigi's schedule
            print(f"{name_etudiant}'s schedule:\n")
            display(student_schedule)
        elif user_choice.upper() == "M":
            cours_ = " "
            while cours_.upper() not in [s.upper() for s in course_list]:
                cours_ = input(f"Please choose a subject from this list:\n{course_list}\n")
                for x in course_list:
                    if cours_.upper() == x.upper():
                        cours_ = x
            print(f"📘 Selected subject: {cours_}")
            
            print("Before Modification ")
            # Select only Luigi's schedule
            student_schedule = schedule_df[name_etudiant]

            # Display Luigi's schedule
            print(f"{name_etudiant}'s schedule:\n")
            display(student_schedule)
            
            print("✏️ Modifying the course...")
            
            
            modify_course(name_etudiant, day_, chosen_time, cours_)
            print("After Modification ")
            
            # Select only Luigi's schedule
            student_schedule = schedule_df[name_etudiant]

            # Display Luigi's schedule
            print(f"{name_etudiant}'s schedule:\n")
            display(student_schedule)
```
